# Comparison of multi-threading benchmark results across modelling PC and Macbook

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mac_results = pd.read_csv('Benchmark_results_mac/benchmark_results.csv')
pc_results = pd.read_csv('Modelling_pc_benchmarks/benchmark_results.csv')

single_thread_pc_time = np.array([74.85 for _ in range(max(len(mac_results['workers']), len(pc_results['workers'])))])
single_thread_mac_time = np.array([59 for _ in range(max(len(mac_results['workers']), len(pc_results['workers'])))])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(mac_results["workers"], mac_results["total_time"], marker='o', label='Multi-processor Mac Time')
plt.plot(pc_results["workers"], pc_results["total_time"], marker='o', label='Multi-processor PC Time')
plt.plot(single_thread_mac_time, linestyle='--', label='Single Process Mac Time')
plt.plot(single_thread_pc_time, linestyle='--', label='Single Process PC Time')
plt.xlabel("Number of Workers")
plt.ylabel("Total Time (s)")
plt.title("Total Processing Time vs Number of Workers")
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(mac_results["workers"], mac_results["samples_per_second"], marker='o', label='Mac')
plt.plot(pc_results["workers"], pc_results["samples_per_second"], marker='o', label='PC')
plt.xlabel("Number of Workers")
plt.ylabel("Samples per Second")
plt.title("Samples per Second vs Number of Workers")
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(mac_results["workers"], mac_results["time_per_worker"], marker='o', label='Mac')
plt.plot(pc_results["workers"], pc_results["time_per_worker"], marker='o', label='PC')
plt.xlabel("Number of Workers")
plt.ylabel("Average Time per Worker (s)")
plt.title("Average Time per Worker vs Number of Workers")
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
